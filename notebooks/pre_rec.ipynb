Precision ( Chnh xc) tr li cu hi: "Trong s nhng ngi m m hnh d on l mc bnh, c bao nhiu ngi thc s mc bnh?", Precision o lng  tin cy ca cc d on dng tnh.
*  ngha trong y t: Precision cao (gn $1.0$) ngha l khi m hnh a ra cnh bo "Vim phi," bn c th rt tin tng vo kt qu .
* Trade-off: Mun Precision cao, m hnh phi rt thn trng khi d on Dng tnh, iu ny thng dn n vic b st nhiu ca bnh (Recall thp).

Recall ( Nhy) (hay cn gi l Sensitivity -  nhy) tr li cu hi: "Trong s nhng ngi thc s mc bnh, m hnh  tm ra c bao nhiu ngi?", Recall o lng kh nng ca m hnh trong vic tm thy tt c cc trng hp dng tnh.
*  ngha trong y t: Recall cao (gn $1.0$) ngha l m hnh rt nhy bn v t b st cc ca bnh thc t. Trong chn on Vim phi, bn mun Recall cao  gim thiu ri ro b st bnh (FN).
* Trade-off: Mun Recall cao, m hnh phi "d di" hn khi d on Dng tnh, iu ny thng dn n nhiu cnh bo sai (Precision thp).

nh i (Trade-off)

Precision v Recall lun c s nh i, c bit l khi bn thay i ngng phn loi (threshold):
* Tng ngng (Threshold $\uparrow$): M hnh tr nn kht khe hn (ch d on Dng tnh khi xc sut rt cao). $\rightarrow$ Precision $\uparrow$, Recall $\downarrow$.
* Gim ngng (Threshold $\downarrow$): M hnh tr nn d di hn (d on Dng tnh ngay c khi xc sut thp). $\rightarrow$ Recall $\uparrow$, Precision $\downarrow$.
